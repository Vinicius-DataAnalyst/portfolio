```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/diogenesjusto/FIAP-On-line-Graduation/main/datasets/arvore_titanic.csv')
df.head()
```

```python
from pycaret.classification import *
s = setup(df, target = 'Survived')
```

```python
best = compare_models()
```

```python
print (best)
```

```python
evaluate_model (best)
```

```python
best_5_models = compare_models (n_select = 5)
```

```python
print(best_5_models)
```

```python
evaluate_model (best_5_models [1])
```
