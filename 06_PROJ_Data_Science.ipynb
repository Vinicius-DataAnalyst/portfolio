Projeto com intenção de aplicar na prática conhecimentos de uso da correlação de variáveis e regressão linear em um conjunto de dados de exemplo.

```python
import pandas as pd
# Análise de correlação linear entre variáveis
df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP-On-Line-Graduation/main/datasets/pib.csv")
df.head()
```

```python
# criando um gráfico para entender o comportamento de 3 colunas
graf = df.iloc[:, 2:5]
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (15,4)

fig, ax = plt.subplots()
graf.plot(ax=ax);
```

```python
# Podemos calcular, então, a matriz de correlação entre essas variáveis para analisarmos os coeficientes de correlação.**
df_corr = df.iloc[ :, 2:15 ]
df_corr.corr()
```

```python
# criando um gráfico de mapa de calor a partir da matrix de correlação
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(df_corr.corr(), annot=True, cmap='YlGnBu', fmt=".2f")
plt.title('Mapa de Calor da Matriz de Correlação')
plt.show()
```

aplicanod modelo de regressão linear

```python
# 1 - Carga das bibliotecas
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
# 2 - Carga dos dados
df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/dados/wine/red_wine4_30.csv", encoding='latin-1')
# 3 - Preparação dos dados
# 4 - Separação das amostras de treino e teste
x_treino, x_teste, y_treino, y_teste = train_test_split(df[['wine_year', 'wine_price']], df['wine_rating'],test_size=0.3266, random_state=33)
# 5 - Construção do modelo
mod = LinearRegression()
mod.fit(x_treino,y_treino)
# 6 - Avaliação do modelo
y_prev = mod.predict(x_teste)
mean_squared_error(y_teste, y_prev)
```

```python
# 1 - Carga das bibliotecas
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
# 2 - Carga dos dados
df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP-On-Line-Graduation/main/datasets/pib.csv")

# 3 - Preparação dos dados

# 4 - Separação das amostras de treino e teste
x_treino, x_teste, y_treino, y_teste = train_test_split(df[['BRP']], df['PIB'],test_size=0.3266, random_state=33)

# 5 - Construção do modelo
mod = LinearRegression()
mod.fit(x_treino,y_treino)

# Cálculo das estatísticas da regressão linear
import statsmodels.api as sm
mod = sm.OLS(y_treino, x_treino).fit()
print(mod.summary())
# 6 - Avaliação do modelo
y_prev = mod.predict(x_teste)
mean_squared_error(y_teste, y_prev)
```
