```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = '/content/drive/MyDrive/01_DATA ANALYTICS/Estudos/PYTHON 2025/exercicios_projetos/Salary_Data.csv'
```

```python
df = pd.read_csv( path )
```

```python
df.shape
```

```python
df.describe().round()
```

```python
# renomeando colunas
df.rename(columns = { 'YearsExperience' : 'AnosExperiencia',
                      'Salary'          : 'Salario'},
          inplace=True )
```

```python
# grafico de nulos
sns.heatmap( df.isnull(), cbar=False)
#Observa-se que não há dados nulos
```

```python
# histograma em linha da renda
plt.title('Frequência acumulada de renda pelo valor do salário')
sns.kdeplot( df['Salario'] );
# Observa-se que a maior parte concentra-se entre 50.000 e 100.000
```

```python
# Histograma da exeriência
plt.title('Frequência acumulada de experiência pelos anos de experiência')
sns.kdeplot( df['AnosExperiencia'] );
```

```python
# Distribuição salario
sns.boxplot( df['Salario'] )
```

```python
# distribuição experiencia
sns.boxplot( df['AnosExperiencia'] )
```

```python
# verificando distribuição de dados com gráfico de dispersão
sns.scatterplot( data=df, x = 'Salario', y= 'AnosExperiencia' );
#Observa-se alta cccorrelação visual entre salario e anos de experiencia
```

```python
sns.regplot( data=df, x = 'Salario', y= 'AnosExperiencia' );
```
